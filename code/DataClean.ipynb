### Data on Games

We have two files for games: games.csv and 2019_games.csv. We will organize and append the datasets with the relevant columns.

```python
games = pd.read_csv('../MLB/games.csv')
```

```python
games
```

```python
col_keeps = ['g_id', 'home_team', 'away_team', 'home_final_score', 'away_final_score']

col_del =[col for col in games.columns if col not in col_keeps]

games.drop(col_del, axis=1, inplace=True)
```

```python
games = games.reindex(columns=col_keeps)
```

```python
games19 = pd.read_csv('../MLB/2019_games.csv')
```

```python
games19
```

```python
### drop irrelevant and Null columns

col_del = [col for col in games19.columns if col not in col_keeps]

games19.drop(col_del, axis=1, inplace=True)
```

```python
### Change data type to be suitable for anaysis

for col in ['g_id','away_final_score', 'home_final_score']:
    games19[col] = games19[col].astype(int)
```

```python
games19.dtypes
```

```python
games = games.append(games19, ignore_index=True)
```

```python
games
```

```python
games.to_csv('../clean_data/clean_games.csv', index=False)
```

### Data on At bats

In this section, we are cleaning the data on at bats.

```python
atbats = pd.read_csv('../MLB/atbats.csv')
atbats19 = pd.read_csv('../MLB/2019_atbats.csv')
```

```python
atbats
```

```python
atbats19
```

```python
### Changing data types to be suitable for analysis

for col in ['inning', 'ab_id', 'g_id']:
    atbats[col] = atbats[col].astype(int)
    atbats19[col] = atbats19[col].astype(int)
```

```python
### Reindex to match so that we can append the DataFrames
col_order = ['ab_id', 'g_id', 'batter_id', 'pitcher_id', 'stand', 'p_throws','inning', 'event', 'o', 'top']

atbats = atbats.reindex(columns=col_order)
atbats19 = atbats19.reindex(columns=col_order)
```

```python
atbats = atbats.append(atbats19, ignore_index=True)
```

```python
atbats.top = atbats.top.astype(bool)
```

```python
atbats
```

```python
atbats.to_csv('../clean_data/clean_atbats.csv', index=False)
```

### Data on pitches

Clean and organize data on individual pitches

```python
pitches = pd.read_csv('../MLB/pitches.csv')
pitches19 = pd.read_csv('../MLB/2019_pitches.csv')
```

```python
pitches
```

```python
pitches19
```

```python
pitches.columns
```

```python
### Dropping all columns that will not be used for analysis

for col in ['start_speed', 'spin_rate', 'spin_dir', 'type_confidence', 'break_angle', 'break_length', 'break_y', 'ax',
       'ay', 'az', 'sz_bot', 'sz_top', 'vx0', 'vy0', 'vz0', 'x', 'x0', 'y',
       'y0', 'z0', 'pfx_x', 'pfx_z', 'nasty', 'zone', 'code', 'type']:
    
    pitches = pitches.drop(col, axis=1)
    pitches19 = pitches19.drop(col, axis=1)
```

```python
### Changing data types for columns 

for col in ['b_score', 'ab_id', 'b_count', 's_count', 'outs', 'pitch_num']:
    pitches[col] = pitches[col].astype(int)
    pitches19[col] = pitches19[col].astype(int)
```

```python
for col in ['on_1b', 'on_2b', 'on_3b']:
    pitches[col] = pitches[col].astype(bool)
    pitches19[col] = pitches19[col].astype(bool)
```

```python
### Now that the two DataFrames have the same data structure, we can append them.

pitches = pitches.append(pitches19, ignore_index=True)
```

```python
pitches
```

```python
pitches.to_csv('../clean_data/clean_pitches.csv', index=False)
```
