In this first section, we will create a data table with the record of each seasonal team.

```python
games = pd.read_csv('../clean_data/clean_games.csv')
```

```python
''' Getting all of the teams in the league. There are 30 teams total in the MLB. Each team plays 162 games
    in a given season. All contests ends with a decision (no ties).'''

teams = games.home_team.unique()
teams
```

```python
''' Each dictionary will hold the record of each team in the given year represented by the two integers at the 
    end of variable name (rec## = records for the year 20##). The keys are the teams and the 1x2 list is the record
    [wins, losses]'''

rec15 = {}
rec16 = {}
rec17 = {}
rec18 = {}
rec19 = {}

for team in teams:
    rec15[team] = [0,0]
    rec16[team] = [0,0]
    rec17[team] = [0,0]
    rec18[team] = [0,0]
    rec19[team] = [0,0]
```

```python
rec19
```

```python
''' This loop will traverse through the row of the dataframe 'games' and will +1 to the win column of the
    team who has a higher score. The other team will +1 to the loss column. The first four numbers of the game_id
    correspond to the year of the game. At the end of the loop, the sum of every team's season's wins and losses 
    will be 162. '''

for game in range(len(games)):
    
    if games.g_id[game] // 100000 == 2015:
        if games.home_final_score[game] > games.away_final_score[game]:
            rec15[games.home_team[game]][0] += 1
            rec15[games.away_team[game]][1] += 1
        else:
            rec15[games.home_team[game]][1] += 1
            rec15[games.away_team[game]][0] += 1
            
    elif games.g_id[game] // 100000 == 2016:
        if games.home_final_score[game] > games.away_final_score[game]:
            rec16[games.home_team[game]][0] += 1
            rec16[games.away_team[game]][1] += 1
        else:
            rec16[games.home_team[game]][1] += 1
            rec16[games.away_team[game]][0] += 1

        
    elif games.g_id[game] // 100000 == 2017:
        if games.home_final_score[game] > games.away_final_score[game]:
            rec17[games.home_team[game]][0] += 1
            rec17[games.away_team[game]][1] += 1
        else:
            rec17[games.home_team[game]][1] += 1
            rec17[games.away_team[game]][0] += 1

        
    elif games.g_id[game] // 100000 == 2018:
        if games.home_final_score[game] > games.away_final_score[game]:
            rec18[games.home_team[game]][0] += 1
            rec18[games.away_team[game]][1] += 1
        else:
            rec18[games.home_team[game]][1] += 1
            rec18[games.away_team[game]][0] += 1
        
    elif games.g_id[game] // 100000 == 2019:
        if games.home_final_score[game] > games.away_final_score[game]:
            rec19[games.home_team[game]][0] += 1
            rec19[games.away_team[game]][1] += 1
        else:
            rec19[games.home_team[game]][1] += 1
            rec19[games.away_team[game]][0] += 1
```

```python
''' We are to take the data we just mined and create a new data frame. The team_id is the unique key that 
    will identify a specific team in a specific year. The first four numbers refer to the year and the last
    two refer to the team'''

records = pd.DataFrame(columns=['team_id', 'team', 'year', 'wins', 'losses'])
```

```python
team_id_m = {}
num = 1

for team in games.home_team.unique():
    team_id_m[team] = num
    num += 1
```

```python
team_id_m
```

```python
curr = 2015

for year in [rec15, rec16, rec17, rec18, rec19]:
    
    for key in year.keys():
        records = records.append({'team_id': ((curr*100) + team_id_m[key]),
                                  'team': key,
                                  'year': curr,
                                  'wins': year[key][0],
                                  'losses': year[key][1]}, ignore_index=True)
        
    curr += 1
```

```python
records['win_%'] = records['wins'] / (records['wins'] + records['losses'])
```
